```python
import cv2
import numpy as np
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
def contour(img,i):
    
    grey = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    grey = cv2.GaussianBlur(grey,(9,9),3)
    edge = cv2.Canny(grey,110,170,L2gradient=True)
    cnts,hier = cv2.findContours(edge,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    for c in cnts:
        perimeter = cv2.arcLength(c,True)
        approx = cv2.approxPolyDP(c,0.02*perimeter,True)
        if len(approx)==4:
            for i in range(4):
                if np.count_nonzero(approx==approx[i])/2 > 1.0:
                    return None
            return approx
```

```python
def ocr(img):
    shape = img.shape[:2]
    #print(shape)
    if shape[0]<=shape[1]:
        x,y=1500,1300
    else:
        x,y=1300,1500
    #print(x,y)
    res = cv2.resize(img, (x,y), interpolation = cv2.INTER_AREA) #1200,1500
    grey_img = cv2.cvtColor(res,cv2.COLOR_BGR2GRAY)
    _,thresh=cv2.threshold(grey_img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    text = pytesseract.image_to_string(grey_img,lang='eng',config='--psm 6')
    #print(text)
    return text
```

```python
def saveFile(text,i):
    with open('myFile'+i+'.txt','w') as file:
        file.write(text)
    file = open('myFile'+i+'.txt','r') 
    for each in file:
        print(each)
    print('---------------------------------------------------')
```

```python
def main():
    img=[]
    for i in range(1,6): #1->4
        i=str(i)
        file = cv2.imread("C:\\Users\\priya\\Desktop\\mini project\\D"+i+".jpg")
        file = cv2.resize(file, (620,620), interpolation = cv2.INTER_AREA)
        cnts = contour(file,i)
        if cnts is not None:
            cnts = np.float32(cnts)
            pts = np.float32([[[0,0],[0,640],[400,640],[400,0]]])
            dest = cv2.getPerspectiveTransform(cnts,pts)
            img.append(cv2.warpPerspective(file,dest,(400,640)))
        if cnts is None:
            roi= cv2.selectROI(file)
            #print(roi)
            img.append(file[(roi[1]):int(roi[1]+roi[3]),int(roi[0]):int(roi[0]+roi[2])])
    for i in range(0,len(img)):
        #print(img[i])
        cv2.imshow('scanned'+str(i),img[i])
        text=ocr(img[i])
        #print(text)
        saveFile(text,str(i))
        #print("***************************************************************")
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if __name__=='__main__':
    main()
```
